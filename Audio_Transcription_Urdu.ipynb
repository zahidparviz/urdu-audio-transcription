```python
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
model = Wav2Vec2ForCTC.from_pretrained("Zaafir/urdu-asr")
tokenizer = Wav2Vec2Tokenizer.from_pretrained("Zaafir/urdu-asr")
```

```python
import os
audio_files = [f for f in os.listdir('wav_files') if f.endswith('.wav')]
```

```python
import speech_recognition as sr
import io
r = sr.Recognizer()
for audio_file in audio_files:
    audio_path = os.path.join('wav_files', audio_file)
    with sr.AudioFile(audio_path) as source:
        audio = r.record(source)
        try:
            text = r.recognize_google(audio, language="ur-IN")
            print(f"Transcribing {audio_file}...")
            text_file = f"{audio_file[:-4]}_transcription.txt"
            text_path = os.path.join('wav_files', text_file)
            with io.open(text_path, "w", encoding="utf-8") as f:
                f.write(text)
        except:
            print(f"Sorry, could not transcribe {audio_file}.")
```
